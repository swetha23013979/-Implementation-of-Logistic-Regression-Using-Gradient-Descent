OUTLIERS DETECTION AND REMOVAL USING IQR

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
swe=[1,3,28,27,25,92,30,39,40,50,26,24,29,94]
ms=pd.DataFrame(swe)
ms
```

```python
sns.boxplot(data=ms)
```

```python
sns.scatterplot(data=ms)
```

```python
q1=ms.quantile(0.25)
q2=ms.quantile(0.05)
q3=ms.quantile(0.75)
IQR=q3-q1
```

```python
IQR
```

```python
q1=np.percentile(ms,25)
q2=np.percentile(ms,50)
q3=np.percentile(ms,75)
IQR=q3-q1
```

```python
IQR
```

```python
low=q1-1.5*IQR
upp=q1+1.5*IQR
```

```python
low
```

```python
upp
```

```python
outliers=[x for x in swe if x<low or x>upp]
```

```python
print("q1 : ",q1)
print("q3 : ",q3)
print("IQR : ",IQR)
print("low : ",low)
print("upp : ",upp)
print("outliers : ",outliers)
```

```python
ms=ms[((ms>=low)&(ms<=upp))]
```

```python
ms
```

```python
ms.dropna()
```

```python
data=[1,2,2,2,3,1,1,15,2,2,2,3,1,1,2]
mean=np.mean(data)
std=np.std(data)
print("MEAN : ",mean)
print("STANDARD DEVIATION : ",std)
```

```python
thresh=3
outlier=[]
for i in data:
  z=(i-mean)/std
  if z > thresh:
    outlier.append(i)
print("outlier in dbs is",outlier)
```
